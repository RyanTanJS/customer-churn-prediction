```python
import pandas as pd,numpy as np,matplotlib.pyplot as plt,seaborn as sns
df = pd.read_csv('../data/processed/telco_churn_clean_v1.csv')
```

```python
changeType = [
    "MultipleLines",
    "InternetService",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies",
    "Contract",
    "PaperlessBilling",
    "PaymentMethod",
    "Churn"]

for col in changeType:
    df[col] = df[col].astype('category')
```

```python

df.info()
df.head()

```

```python
categories = [
    "MultipleLines",
    "InternetService",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies",
    "Contract",
    "PaperlessBilling",
    "PaymentMethod",
    "Churn"]


for col in categories:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col,data=df)
    plt.title(f'distribution of {col}')
    plt.show()
```
