```python
import pandas as pd,numpy as np,matplotlib.pyplot as plt,seaborn as sns
df = pd.read_csv('../data/processed/telco_churn_clean_v1.csv')
```

```python
changeType = [
    "MultipleLines",
    "InternetService",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies",
    "Contract",
    "PaperlessBilling",
    "PaymentMethod",
    "Churn"]

for col in changeType:
    df[col] = df[col].astype('category')
```

```python
df.info()
df.head()
```

```python
categories = [
    "MultipleLines",
    "InternetService",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies",
    "Contract",
    "PaperlessBilling",
    "PaymentMethod",
    "Churn"]


for col in categories:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col,data=df)
    plt.title(f'distribution of {col}')
    plt.show()
```

```python
numericCategories=[
    'tenure',
    'MonthlyCharges',
    'TotalCharges']

for col in numericCategories:
    plt.figure(figsize=(6,4))
    sns.histplot(x=col,data=df)
    plt.title(f'distribution of {col}')
    plt.show()
```

```python
for col in numericCategories:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=col,data=df)
    plt.title(f'distribution of {col}')
    plt.plot()
```

bivariate analysis

```python
sns.countplot(x='Contract',hue='Churn',data=df)
plt.title('Churn rate by contract type')
plt.plot()
```

```python
sns.countplot(x='TechSupport',hue='Churn',data=df)
plt.title('Churn rate by tech support type')
plt.plot()
```

```python
sns.countplot(x='OnlineSecurity',hue='Churn',data=df)
plt.title('Churn rate by security type')
plt.plot()
```

```python
sns.boxplot(x='MonthlyCharges',hue='Churn',data=df)
plt.title('Churn rate by monthly charge amount')
plt.plot()
```

```python
sns.boxplot(x='tenure',hue='Churn',data=df)
plt.title('Churn rate by tenure')
plt.plot()
```
